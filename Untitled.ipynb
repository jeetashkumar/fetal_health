```python
corr = data.corr()

plt.figure(figsize=(15,10))
sns.heatmap(corr, annot=True)
plt.show()
```

```python
def search_strong_relation(value):
    if value >= 0.7:
        return value
    elif value<=-0.7:
        return value 
    else:
        return 0
```

```python
corr = data.corr()
corr = (corr>=0.7).astype('int')

plt.figure(figsize=(15,10))
plt.title("effective Positive Relation")
sns.heatmap(corr, annot=True)
plt.show()
```

```python
corr = data.corr()
corr = (corr<-0.7).astype('int')

plt.figure(figsize=(15,10))
plt.title("effective Negative Relation")
sns.heatmap(corr, annot=True)
plt.show()
```
